```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('weather_test.csv')
```

```python
data
```

```python
data.columns
```

```python
data1=data.drop([' _fog',' _hail', ' _heatindexm',' _precipm',' _thunder',' _tornado',' _wgustm',' _windchillm',' _vism', ' _wdird', ' _wdire',' _wspdm',' _tempm'], axis=1)
```

```python
data1
```

```python
data1.dropna(inplace=True)
```

```python
data1
```

```python
print(sns.pairplot(data1,hue=' _conds'))
```

```python
from sklearn.model_selection import train_test_split
```

```python
data1.columns
```

```python
data1['date']=data1['datetime_utc'].str.split('-').str.get(0)
data1['date']
data1['time']=data1['datetime_utc'].str.split('-').str.get(1)
data1['time']=data1['time'].str.split(':').str.get(0)
data1['time']
```

```python
X=data1[[' _dewptm', ' _hum', ' _pressurem', ' _rain',
       ' _snow','date','time']]
y=data1[' _conds']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
ob=KNeighborsClassifier()
```

```python
ob.fit(X_train,y_train)
```

```python
prediction=ob.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    ob = KNeighborsClassifier(n_neighbors=i)
    ob.fit(X_train,y_train)
    pred_i = ob.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
ob = KNeighborsClassifier(n_neighbors=30)

ob.fit(X_train,y_train)
prediction = ob.predict(X_test)

print('WITH K=30')
print('\n')
print(classification_report(y_test,prediction))
```
